```python
import re

def find_valid_multiplications(text):
    pattern = r'mul\((\d{1,3}),(\d{1,3})\)'

    # Find Matches in Text
    matches = re.finditer(pattern, text)

    # Process Each Match into (x, y) Tuples
    multiplications = []
    for match in matches:
        x = int(match.group(1))
        y = int(match.group(2))
        multiplications.append((x, y))

    return multiplications

def calculate_total(multiplications):
    total = 0
    for x, y in multiplications:
        total += x * y
    return total

# Input File
with open('day3.txt', 'r') as f:
    corrupted_memory = f.read()

# Find Valid Multiplications
valid_muls = find_valid_multiplications(corrupted_memory)

# Calculate Total
result = calculate_total(valid_muls)

print(f"Total sum of all multiplication results: {result}")
```

```python
import re

def find_control_and_mul_instructions(text):
    # Pattern for Valid mul(X,Y) Where X and Y are 1-3 Digit Numbers
    mul_pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
    # Pattern for Control Instructions
    do_pattern = r'do\(\)'
    dont_pattern = r'don\'t\(\)'

    instructions = []

    # Pattern Matches with Their Positions
    mul_matches = [(m.start(), 'mul', m.groups()) for m in re.finditer(mul_pattern, text)]
    do_matches = [(m.start(), 'do', None) for m in re.finditer(do_pattern, text)]
    dont_matches = [(m.start(), 'dont', None) for m in re.finditer(dont_pattern, text)]

    # Combine Matches and Sort by Position
    all_matches = mul_matches + do_matches + dont_matches
    all_matches.sort(key=lambda x: x[0])

    # Convert Format
    for _, instr_type, values in all_matches:
        if instr_type == 'mul':
            x, y = map(int, values)
            instructions.append(('mul', (x, y)))
        else:
            instructions.append((instr_type, None))

    return instructions

def calculate_enabled_multiplications(instructions):
    total = 0
    enabled = True

    for instr_type, values in instructions:
        if instr_type == 'do':
            enabled = True
        elif instr_type == 'dont':
            enabled = False
        elif instr_type == 'mul' and enabled:
            x, y = values
            total += x * y

    return total

# Input File
with open('day3.txt', 'r') as f:
    corrupted_memory = f.read()

# Find Instructions
instructions = find_control_and_mul_instructions(corrupted_memory)

# Total of Enabled Multiplications
result = calculate_enabled_multiplications(instructions)

print(f"Sum of Enabled Multiplication Results: {result}")
```

```python
def find_xmas_occurrences(grid):
    rows = len(grid)
    cols = len(grid[0])
    count = 0

    # Eight Directions
    directions = [
        (0, 1),   # right
        (1, 0),   # down
        (1, 1),   # diagonal down-right
        (1, -1),  # diagonal down-left
        (0, -1),  # left
        (-1, 0),  # up
        (-1, 1),  # diagonal up-right
        (-1, -1)  # diagonal up-left
    ]

    def check_xmas(row, col, dr, dc):
        if not (0 <= row + 3*dr < rows and 0 <= col + 3*dc < cols):
            return False

        return (grid[row][col] == 'X' and
                grid[row + dr][col + dc] == 'M' and
                grid[row + 2*dr][col + 2*dc] == 'A' and
                grid[row + 3*dr][col + 3*dc] == 'S')

    # Check Each Starting Position in Each Direction
    for row in range(rows):
        for col in range(cols):
            for dr, dc in directions:
                if check_xmas(row, col, dr, dc):
                    count += 1

    return count

# Input File
with open('day4.txt', 'r') as f:
    grid = f.read().strip().split('\n')

# Total Occurrences
result = find_xmas_occurrences(grid)
print(f"Total occurrences of 'XMAS': {result}")
```

```python
from collections import defaultdict, deque

def parse_input(input_text):
    lines = [line.strip() for line in input_text.split('\n') if line.strip()]
    rules = defaultdict(set)
    updates = []

    for line in lines:
        if '|' in line:
            before, after = map(int, line.split('|'))
            rules[before].add(after)
        else:
            update = list(map(int, line.split(',')))
            updates.append(update)

    return rules, updates

def is_valid_order(pages, rules):
    for i in range(len(pages)):
        for j in range(i + 1, len(pages)):
            page1 = pages[i]
            page2 = pages[j]
            if page2 in rules and page1 in rules[page2]:
                return False
    return True

def build_graph(pages, rules):
    # Adjacency List Representation
    graph = defaultdict(set)
    in_degree = defaultdict(int)

    for page in pages:
        if page not in graph:
            graph[page] = set()

    # Add Edges Based on Rules (A->B means A must come before B)
    for page1 in pages:
        for page2 in pages:
            if page1 in rules and page2 in rules[page1]:
                graph[page1].add(page2)
                in_degree[page2] += 1

    return graph, in_degree

def topological_sort(pages, rules):
    # Build Graph
    graph, in_degree = build_graph(pages, rules)

    # Nodes That Have No Prerequisites
    queue = deque([page for page in pages if in_degree[page] == 0])
    result = []

    # Process Nodes
    while queue:
        # Get Next Node with no Prerequisites
        page = queue.popleft()
        result.append(page)

        # Process Neighbors
        for next_page in list(graph[page]):
            in_degree[next_page] -= 1
            if in_degree[next_page] == 0:
                queue.append(next_page)

    if len(result) != len(pages):
        return None

    return result

def get_middle_number(sequence):
    return sequence[len(sequence) // 2]

# Input File
with open('day5.txt', 'r') as f:
    input_text = f.read()

# Parse Input
rules, updates = parse_input(input_text)

# Process Incorrect Updates
middle_numbers = []
incorrect_updates = []

# First identify incorrect updates
for update in updates:
    if not is_valid_order(update, rules):
        incorrect_updates.append(update)

# Sort and Process Incorrect Updates
for update in incorrect_updates:
    sorted_update = topological_sort(update, rules)
    if sorted_update:
        middle = get_middle_number(sorted_update)
        middle_numbers.append(middle)
        print(f"Original: {update}")
        print(f"Sorted  : {sorted_update}")
        print(f"Middle  : {middle}\n")

# Calculate Sum
result = sum(middle_numbers)
print(f"Sum of middle numbers from corrected updates: {result}")
print(f"Number of incorrect updates: {len(middle_numbers)}")
print(f"Middle numbers: {middle_numbers}")
```

```python
def load_map(filepath):
    with open(filepath, 'r') as file:
        return [list(line.strip()) for line in file.readlines()]

def get_start_position(grid):
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            if grid[y][x] == '^':
                return (x, y, 0)
    return None

def simulate_movement_no_obstacle(grid):
    dx = [0, 1, 0, -1]
    dy = [-1, 0, 1, 0]

    x, y, direction = get_start_position(grid)
    path = set()

    while True:
        next_x = x + dx[direction]
        next_y = y + dy[direction]

        if not (0 <= next_y < len(grid) and 0 <= next_x < len(grid[0])):
            break

        if grid[next_y][next_x] == '#':
            direction = (direction + 1) % 4
        else:
            x, y = next_x, next_y
            path.add((x, y))

    return path

def check_loop(grid, obstacle_x, obstacle_y, original_path):
    dx = [0, 1, 0, -1]
    dy = [-1, 0, 1, 0]

    x, y, direction = get_start_position(grid)
    states = set()
    states.add((x, y, direction))

    steps = 0
    max_steps = len(original_path) * 4

    # Early exit if obstacle isn't near the original path
    if (obstacle_x, obstacle_y) not in original_path:
        near_path = False
        for dx_check, dy_check in [(0,1), (1,0), (0,-1), (-1,0)]:
            check_x, check_y = obstacle_x + dx_check, obstacle_y + dy_check
            if (check_x, check_y) in original_path:
                near_path = True
                break
        if not near_path:
            return False

    while steps < max_steps:
        next_x = x + dx[direction]
        next_y = y + dy[direction]

        if not (0 <= next_y < len(grid) and 0 <= next_x < len(grid[0])):
            return False

        if grid[next_y][next_x] == '#' or (next_x == obstacle_x and next_y == obstacle_y):
            direction = (direction + 1) % 4
        else:
            x, y = next_x, next_y

        state = (x, y, direction)
        if state in states:
            return True
        states.add(state)
        steps += 1

    return steps == max_steps

def find_loop_positions(grid):
    start_pos = get_start_position(grid)
    if not start_pos:
        return 0

    # Get the original path
    original_path = simulate_movement_no_obstacle(grid)
    loop_count = 0

    for y in range(len(grid)):
        for x in range(len(grid[0])):
            if grid[y][x] == '.' and (x, y) != (start_pos[0], start_pos[1]):
                if check_loop(grid, x, y, original_path):
                    loop_count += 1

    return loop_count

def main():
    grid = load_map('day6.txt')
    result = find_loop_positions(grid)
    print(f"There are {result} positions where placing an obstruction would create a loop.")

if __name__ == "__main__":
    main()
```

```python
from itertools import product

def evaluate_expression(numbers, operators):
    result = numbers[0]
    for i, op in enumerate(operators):
        if op == '+':
            result += numbers[i + 1]
        else:  # op == '*'
            result *= numbers[i + 1]
    return result

def parse_line(line):
    test_val, numbers = line.strip().split(': ')
    return int(test_val), [int(x) for x in numbers.split()]

def can_form_test_value(test_val, numbers):
    # For n numbers, we need n-1 operators
    num_operators = len(numbers) - 1
    if num_operators == 0:
        return test_val == numbers[0]

    # Try all possible combinations of operators
    for ops in product(['+', '*'], repeat=num_operators):
        if evaluate_expression(numbers, ops) == test_val:
            return True
    return False

def solve_calibration(filepath):
    total = 0

    with open(filepath, 'r') as file:
        for line in file:
            if not line.strip():
                continue
            test_val, numbers = parse_line(line)
            if can_form_test_value(test_val, numbers):
                total += test_val

    return total

# Run the solution
result = solve_calibration('day7.txt')
print(f"Total calibration result: {result}")
```

```python
from itertools import product

def concat_numbers(a, b):
    if b == 0:
        return a * 10
    b_length = len(str(b))
    return a * (10 ** b_length) + b

def evaluate_expression(numbers, operators):
    result = numbers[0]
    for i, op in enumerate(operators):
        if op == '+':
            result += numbers[i + 1]
        elif op == '*':
            result *= numbers[i + 1]
        else:  # op == '||'
            result = concat_numbers(result, numbers[i + 1])
    return result

def parse_line(line):
    test_val, numbers = line.strip().split(': ')
    return int(test_val), [int(x) for x in numbers.split()]

def can_form_test_value(test_val, numbers):
    # For n numbers, we need n-1 operators
    num_operators = len(numbers) - 1
    if num_operators == 0:
        return test_val == numbers[0]

    # Try all possible combinations of operators
    for ops in product(['+', '*', '||'], repeat=num_operators):
        try:
            if evaluate_expression(numbers, ops) == test_val:
                return True
        except (OverflowError, ValueError):
            # Skip combinations that would cause overflow
            continue
    return False

def solve_calibration_part2(filepath):
    total = 0

    with open(filepath, 'r') as file:
        for line in file:
            if not line.strip():
                continue
            test_val, numbers = parse_line(line)
            if can_form_test_value(test_val, numbers):
                total += test_val

    return total

# Run the solution
result = solve_calibration_part2('day7.txt')
print(f"Total calibration result (Part 2): {result}")
```

```python
def load_map(filepath):
    with open(filepath, 'r') as f:
        grid = [list(line.strip()) for line in f]

    height = len(grid)
    width = len(grid[0])
    antennas = {}  # frequency -> list of (x, y) coordinates

    for y in range(height):
        for x in range(width):
            if grid[y][x] != '.':
                freq = grid[y][x]
                if freq not in antennas:
                    antennas[freq] = []
                antennas[freq].append((x, y))

    return antennas, width, height

def is_collinear(p1, p2, p3):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    # Cross product should be zero for collinear points
    return (y2 - y1) * (x3 - x1) == (y3 - y1) * (x2 - x1)

def check_distances(point, ant1, ant2):
    x, y = point
    x1, y1 = ant1
    x2, y2 = ant2

    # Calculate squared distances to avoid floating point errors
    d1_sq = (x - x1)**2 + (y - y1)**2
    d2_sq = (x - x2)**2 + (y - y2)**2

    # Skip if point coincides with either antenna
    if d1_sq == 0 or d2_sq == 0:
        return False

    # Check if one distance is twice the other
    # Using squared distances means one should be 4 times the other
    return abs(d1_sq - 4*d2_sq) < 0.001 or abs(d2_sq - 4*d1_sq) < 0.001

def find_antinodes(ant1, ant2, width, height):
    antinodes = set()
    x1, y1 = ant1
    x2, y2 = ant2

    # Define the search range
    x_min = max(0, min(x1, x2) - max(width, height))
    x_max = min(width, max(x1, x2) + max(width, height))
    y_min = max(0, min(y1, y2) - max(width, height))
    y_max = min(height, max(y1, y2) + max(width, height))

    # For each point in the bounded area
    for y in range(y_min, y_max):
        for x in range(x_min, x_max):
            point = (x, y)
            # Skip if point is not collinear or doesn't satisfy distance ratio
            if is_collinear(point, ant1, ant2) and check_distances(point, ant1, ant2):
                antinodes.add(point)

    return antinodes

def solve_antinodes(filepath):
    antennas, width, height = load_map(filepath)
    all_antinodes = set()

    # For each frequency
    for freq, positions in antennas.items():
        # For each pair of antennas with same frequency
        for i, ant1 in enumerate(positions):
            for ant2 in positions[i+1:]:
                # Find antinodes for this pair
                new_antinodes = find_antinodes(ant1, ant2, width, height)
                all_antinodes.update(new_antinodes)

    return len(all_antinodes)

# Run the solution
result = solve_antinodes('day8.txt')
print(f"Number of unique antinode locations: {result}")
```

```python
def parse_disk_map(filepath):
    with open(filepath, 'r') as f:
        disk_map = f.read().strip()

    # Parse the alternating sequence into files and spaces
    files = []  # (length, id) tuples
    spaces = []  # just lengths
    file_id = 0

    # Parse alternating digits as file lengths and space lengths
    for i in range(0, len(disk_map), 2):
        file_len = int(disk_map[i])
        files.append((file_len, file_id))
        file_id += 1

        if i + 1 < len(disk_map):
            space_len = int(disk_map[i + 1])
            spaces.append(space_len)

    return files, spaces

def compact_disk(files, spaces):
    # Convert to a list of blocks where each block is (position, length, file_id)
    blocks = []
    current_pos = 0

    # Build initial block layout
    for i, ((file_len, file_id), space_len) in enumerate(zip(files, spaces)):
        blocks.append((current_pos, file_len, file_id))
        current_pos += file_len + space_len

    # Add the last file if it exists (no trailing space)
    if len(files) > len(spaces):
        file_len, file_id = files[-1]
        blocks.append((current_pos, file_len, file_id))

    # Process files from right to left
    for i in range(len(blocks) - 1, 0, -1):
        current_pos, length, file_id = blocks[i]
        target_pos = blocks[i-1][0] + blocks[i-1][1]

        # Move block to fill gap
        if target_pos < current_pos:
            blocks[i] = (target_pos, length, file_id)

    return blocks

def calculate_checksum(final_blocks):
    checksum = 0

    for pos, length, file_id in final_blocks:
        # For each position in the block
        for p in range(pos, pos + length):
            checksum += p * file_id

    return checksum

def solve_disk_compaction(filepath):
    # Parse input
    files, spaces = parse_disk_map(filepath)

    # Simulate compaction
    final_blocks = compact_disk(files, spaces)

    # Calculate checksum
    return calculate_checksum(final_blocks)

# Run the solution
result = solve_disk_compaction('day9.txt')
print(f"Filesystem checksum after compaction: {result}")
```

```python
def solve_compaction(filepath):
    # Read input
    with open(filepath, 'r') as f:
        disk_map = f.read().strip()

    # Initialize disk as dictionary
    disk = {}
    max_write_location = 0

    # Parse input and write initial disk state
    numbers = [int(c) for c in disk_map]
    for file_id in range(len(numbers) // 2 + len(numbers) % 2):
        # Get file size and gap size (if it exists)
        i = file_id * 2
        file_size = numbers[i]
        gap_size = numbers[i + 1] if i + 1 < len(numbers) else 0

        # Write file blocks
        for pointer in range(max_write_location, max_write_location + file_size):
            disk[pointer] = file_id

        # Move pointer past gap
        max_write_location += file_size + gap_size

    # Compact the disk
    writer = 0  # Points to lowest gap
    reader = max_write_location - 1  # Points to highest filled position

    while writer < reader:
        # Find next gap
        while writer in disk and writer < reader:
            writer += 1

        # Find next filled position from the right
        while reader not in disk and reader > writer:
            reader -= 1

        # If pointers crossed, we're done
        if writer >= reader:
            break

        # Move block from reader to writer
        disk[writer] = disk[reader]
        del disk[reader]

    # Calculate checksum
    checksum = 0
    for idx in range(max_write_location):
        if idx in disk:
            checksum += idx * disk[idx]

    return checksum

# Run solution
result = solve_compaction('day9.txt')
print(f"Checksum: {result}")

# Verify with example
example = "2333133121414131402"
with open('example.txt', 'w') as f:
    f.write(example)
example_result = solve_compaction('example.txt')
print(f"\nExample case result: {example_result}")  # Should be 1928

# Debug information
def print_debug_info(filepath):
    with open(filepath, 'r') as f:
        disk_map = f.read().strip()
    print(f"\nInput length: {len(disk_map)}")
    print(f"First few digits: {disk_map[:20]}")

    # Show first few blocks
    disk = {}
    max_pos = 0
    numbers = [int(c) for c in disk_map]
    for i in range(3):  # Show first 3 files
        file_size = numbers[i*2]
        gap_size = numbers[i*2 + 1] if i*2 + 1 < len(numbers) else 0
        print(f"File {i}: size={file_size}, gap={gap_size}, starts at {max_pos}")
        max_pos += file_size + gap_size

print_debug_info('day9.txt')
```

```python
def build_initial_state(disk_map):
    disk = {}  # position -> file_id
    files = []  # list of (start_pos, length, file_id)
    pos = 0

    numbers = [int(c) for c in disk_map]
    for file_id in range(len(numbers) // 2 + len(numbers) % 2):
        i = file_id * 2
        file_size = numbers[i]
        gap_size = numbers[i + 1] if i + 1 < len(numbers) else 0

        # Record file information
        files.append((pos, file_size, file_id))

        # Write file blocks
        for pointer in range(pos, pos + file_size):
            disk[pointer] = file_id

        pos += file_size + gap_size

    return disk, files, pos  # Return max_pos as well

def find_leftmost_space(disk, size, max_pos, start_pos):
    current_gap_start = None
    current_gap_size = 0

    for pos in range(max_pos):
        if pos >= start_pos:
            break

        if pos not in disk:
            if current_gap_start is None:
                current_gap_start = pos
            current_gap_size += 1
            if current_gap_size >= size:
                return current_gap_start
        else:
            current_gap_start = None
            current_gap_size = 0

    return None

def move_file(disk, start_pos, size, new_pos, file_id):
    # Remove old positions
    for pos in range(start_pos, start_pos + size):
        del disk[pos]

    # Write new positions
    for pos in range(new_pos, new_pos + size):
        disk[pos] = file_id

def solve_compaction_part2(filepath):
    # Read input
    with open(filepath, 'r') as f:
        disk_map = f.read().strip()

    # Build initial state
    disk, files, max_pos = build_initial_state(disk_map)

    # Sort files by ID in descending order
    files.sort(key=lambda x: x[2], reverse=True)

    # Try to move each file
    for start_pos, size, file_id in files:
        # Find leftmost space that can fit this file
        new_pos = find_leftmost_space(disk, size, max_pos, start_pos)

        # If we found a valid position and it's to the left of current position
        if new_pos is not None and new_pos < start_pos:
            move_file(disk, start_pos, size, new_pos, file_id)

    # Calculate checksum
    checksum = 0
    for pos in range(max_pos):
        if pos in disk:
            checksum += pos * disk[pos]

    return checksum

# Run solution
result = solve_compaction_part2('day9.txt')
print(f"Part 2 Checksum: {result}")

# Verify with example
example = "2333133121414131402"
with open('example.txt', 'w') as f:
    f.write(example)
example_result = solve_compaction_part2('example.txt')
print(f"\nExample case result: {example_result}")  # Should be 2858
```

```python
from typing import List

def transform_stone(stone: int) -> List[int]:
    # Rule 1: If stone is 0, replace with 1
    if stone == 0:
        return [1]

    # Convert to string to check number of digits
    stone_str = str(stone)

    # Rule 2: If even number of digits, split into two stones
    if len(stone_str) % 2 == 0:
        mid = len(stone_str) // 2
        left = int(stone_str[:mid])
        right = int(stone_str[mid:])
        return [left, right]

    # Rule 3: Multiply by 2024
    return [stone * 2024]

def simulate_blink(stones: List[int]) -> List[int]:
    new_stones = []
    for stone in stones:
        new_stones.extend(transform_stone(stone))
    return new_stones

def count_stones_after_blinks(initial_stones: List[int], blinks: int) -> int:

    current_stones = initial_stones.copy()

    for _ in range(blinks):
        current_stones = simulate_blink(current_stones)

    return len(current_stones)

def solve_plutonian_stones(input_data: str, num_blinks: int = 25) -> int:
    # Parse input string into list of integers
    initial_stones = [int(x) for x in input_data.strip().split()]

    return count_stones_after_blinks(initial_stones, num_blinks)

# Solve
with open('day11.txt', 'r') as file:
    input_data = file.read()

result = solve_plutonian_stones(input_data)
print(f"After 25 blinks, there will be {result} stones.")
```

```python
from typing import List

def count_stones_for_value(stone: int, remaining_blinks: int, cache: dict) -> int:
    if remaining_blinks == 0:
        return 1

    cache_key = (stone, remaining_blinks)
    if cache_key in cache:
        return cache[cache_key]

    # Apply rules
    if stone == 0:
        # 0 becomes 1
        result = count_stones_for_value(1, remaining_blinks - 1, cache)
    elif len(str(stone)) % 2 == 0:
        # Split into two stones
        stone_str = str(stone)
        mid = len(stone_str) // 2
        left = int(stone_str[:mid])
        right = int(stone_str[mid:])
        result = (count_stones_for_value(left, remaining_blinks - 1, cache) +
                 count_stones_for_value(right, remaining_blinks - 1, cache))
    else:
        # Multiply by 2024
        result = count_stones_for_value(stone * 2024, remaining_blinks - 1, cache)

    cache[cache_key] = result
    return result

def solve_plutonian_stones_optimized(input_data: str, num_blinks: int = 75) -> int:
    initial_stones = [int(x) for x in input_data.strip().split()]
    cache = {}

    total_stones = 0
    for stone in initial_stones:
        total_stones += count_stones_for_value(stone, num_blinks, cache)

    return total_stones

# Solve
with open('day11.txt', 'r') as file:
    input_data = file.read()

result = solve_plutonian_stones_optimized(input_data)
print(f"After 75 blinks, there will be {result} stones.")
```

```python
from typing import List, Set, Tuple, Dict
from collections import deque

class Grid:
    def __init__(self, data: List[List[str]]):
        self.grid = data
        self.rows = len(data)
        self.cols = len(data[0])

    def get_neighbors(self, x: int, y: int) -> List[Tuple[int, int]]:
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        neighbors = []

        for dx, dy in directions:
            new_x, new_y = x + dx, y + dy
            if 0 <= new_x < self.rows and 0 <= new_y < self.cols:
                neighbors.append((new_x, new_y))
            else:
                # Count boundary edge
                neighbors.append(None)

        return neighbors

    def is_valid_neighbor(self, pos: Tuple[int, int], plant_type: str) -> bool:
        if pos is None:
            return False
        x, y = pos
        return self.grid[x][y] == plant_type

    def find_region(self, start_x: int, start_y: int, visited: Set[Tuple[int, int]]) -> Tuple[int, int]:
        if (start_x, start_y) in visited:
            return 0, 0

        plant_type = self.grid[start_x][start_y]
        region = set()
        perimeter = 0
        queue = deque([(start_x, start_y)])
        region.add((start_x, start_y))
        visited.add((start_x, start_y))

        while queue:
            x, y = queue.popleft()
            neighbors = self.get_neighbors(x, y)

            # Count perimeter contributions for this plot
            for neighbor in neighbors:
                if not self.is_valid_neighbor(neighbor, plant_type):
                    perimeter += 1
                elif neighbor not in region:
                    nx, ny = neighbor
                    queue.append((nx, ny))
                    region.add(neighbor)
                    visited.add(neighbor)

        return len(region), perimeter

def parse_map(input_data: str) -> List[List[str]]:
    return [list(line.strip()) for line in input_data.strip().splitlines()]

def solve_garden_fencing(input_data: str) -> int:
    garden = Grid(parse_map(input_data))
    visited = set()
    total_price = 0

    # Find Regions
    for i in range(garden.rows):
        for j in range(garden.cols):
            area, perimeter = garden.find_region(i, j, visited)
            if area > 0:  # If we found a new region
                price = area * perimeter
                total_price += price

    return total_price

# Solve
with open('day12.txt', 'r') as file:
    input_data = file.read()

result = solve_garden_fencing(input_data)
print(f"The total price of fencing all regions is: {result}")
```

```python
from typing import List, Set, Tuple, Dict
from collections import deque, defaultdict

DIRECTIONS = [(0, -1), (1, 0), (0, 1), (-1, 0)]  # N, E, S, W
DIR_NAMES = ['N', 'E', 'S', 'W']

class Grid:
    def __init__(self, data: List[List[str]]):
        self.grid = data
        self.rows = len(data)
        self.cols = len(data[0])

    def is_valid(self, x: int, y: int, plant_type: str) -> bool:
        return (0 <= x < self.rows and
                0 <= y < self.cols and
                self.grid[x][y] == plant_type)

    def flood_fill_region(self, start_x: int, start_y: int, visited: Set[Tuple[int, int]]) -> Tuple[int, Set[Tuple[int, int]], Dict[str, Set[Tuple[int, int]]]]:
        if (start_x, start_y) in visited:
            return 0, set(), {}

        plant_type = self.grid[start_x][start_y]
        region = set()
        perimeter_dir = defaultdict(set)

        queue = deque([(start_x, start_y)])
        region.add((start_x, start_y))
        visited.add((start_x, start_y))

        while queue:
            x, y = queue.popleft()

            # Check each direction
            for dir_idx, (dx, dy) in enumerate(DIRECTIONS):
                next_x, next_y = x + dx, y + dy

                # If neighbor is invalid, current plot is part of perimeter in this direction
                if not self.is_valid(next_x, next_y, plant_type):
                    perimeter_dir[DIR_NAMES[dir_idx]].add((x, y))
                else:
                    if (next_x, next_y) not in region:
                        queue.append((next_x, next_y))
                        region.add((next_x, next_y))
                        visited.add((next_x, next_y))

        return len(region), region, perimeter_dir

    def count_continuous_sides(self, perimeter_plots: Set[Tuple[int, int]], direction: str) -> int:
        if not perimeter_plots:
            return 0

        sides = 0
        visited = set()

        # For each unvisited perimeter plot
        for plot in perimeter_plots:
            if plot in visited:
                continue

            # Start a new continuous side
            queue = deque([plot])
            visited.add(plot)

            while queue:
                x, y = queue.popleft()

                # Check adjacent plots in the same direction
                for dx, dy in [(-1, 0), (1, 0)] if direction in 'NS' else [(0, -1), (0, 1)]:
                    next_plot = (x + dx, y + dy)
                    if next_plot in perimeter_plots and next_plot not in visited:
                        queue.append(next_plot)
                        visited.add(next_plot)

            sides += 1

        return sides

    def calculate_sides(self, perimeter_by_direction: Dict[str, Set[Tuple[int, int]]]) -> int:
        total_sides = 0

        for direction, plots in perimeter_by_direction.items():
            sides = self.count_continuous_sides(plots, direction)
            total_sides += sides

        return total_sides

def solve_garden_fencing_part2(input_data: str) -> int:
    # Parse input into grid
    grid = Grid([list(line.strip()) for line in input_data.strip().splitlines()])
    visited = set()
    total_price = 0

    # Process each unvisited position
    for i in range(grid.rows):
        for j in range(grid.cols):
            area, region, perimeter_dir = grid.flood_fill_region(i, j, visited)
            if area > 0:
                sides = grid.calculate_sides(perimeter_dir)
                price = area * sides
                total_price += price

    return total_price

# Solve
with open('day12.txt', 'r') as file:
    input_data = file.read()

result = solve_garden_fencing_part2(input_data)
print(f"The total price of fencing all regions using the number of sides is: {result}")
```

```python
from re import findall

def solve_claw_machines(input_data: str) -> int:
    # Split input into separate machine cases
    machines = input_data.strip().split('\n\n')

    total_tokens = 0
    for machine in machines:
        # Extract all numbers from the machine description
        ax, ay, bx, by, px, py = map(int, findall(r"\d+", machine))

        try:
            # Solve for 'a' using the ratio method
            # From: ax*a + bx*b = px and ay*a + by*b = py
            # Eliminate b to solve for a
            a = round((py/by - px/bx)/(ay/by - ax/bx))

            # Back-solve for b using the x-equation
            b = round((px - a*ax)/bx)

            # Verify that we found exact integer solutions that satisfy both equations
            if (a*ax + b*bx == px and
                a*ay + b*by == py and
                a >= 0 and b >= 0):
                # Calculate token cost: 3 tokens for A, 1 token for B
                total_tokens += 3*a + b

        except (ZeroDivisionError, ValueError):
            # Skip cases where we can't find a solution
            continue

    return total_tokens

# Solve
with open('day13.txt', 'r') as file:
    input_data = file.read()

result = solve_claw_machines(input_data)
print(f"The minimum number of tokens needed to win all possible prizes is: {result}")
```

```python
from dataclasses import dataclass
from typing import List, Tuple
import re

@dataclass
class Robot:
    x: int
    y: int
    vx: int
    vy: int

    def move(self, width: int, height: int) -> None:
        # Update position with wrapping
        self.x = (self.x + self.vx) % width
        self.y = (self.y + self.vy) % height

def parse_input(input_data: str) -> List[Robot]:
    robots = []
    pattern = r'p=(\d+),(\d+) v=(-?\d+),(-?\d+)'

    for line in input_data.strip().split('\n'):
        if match := re.match(pattern, line):
            x, y, vx, vy = map(int, match.groups())
            robots.append(Robot(x, y, vx, vy))

    return robots

def simulate_robots(robots: List[Robot], width: int, height: int, seconds: int) -> List[Robot]:
    # Create new list to avoid modifying original
    current_robots = [Robot(r.x, r.y, r.vx, r.vy) for r in robots]

    # Simulate for specified number of seconds
    for _ in range(seconds):
        for robot in current_robots:
            robot.move(width, height)

    return current_robots

def count_robots_in_quadrants(robots: List[Robot], width: int, height: int) -> Tuple[int, int, int, int]:
    # Initialize quadrant counts
    quadrants = [0, 0, 0, 0]  # [top-left, top-right, bottom-left, bottom-right]

    # Define quadrant boundaries (excluding middle lines)
    mid_x = width // 2
    mid_y = height // 2

    for robot in robots:
        # Skip robots on middle lines
        if robot.x == mid_x or robot.y == mid_y:
            continue

        # Determine quadrant
        quadrant_idx = (
            (1 if robot.x > mid_x else 0) +
            (2 if robot.y > mid_y else 0)
        )
        quadrants[quadrant_idx] += 1

    return tuple(quadrants)

def calculate_safety_factor(robots: List[Robot], width: int, height: int) -> int:
    q1, q2, q3, q4 = count_robots_in_quadrants(robots, width, height)
    return q1 * q2 * q3 * q4

def solve_robot_safety(input_data: str, width: int = 101, height: int = 103, seconds: int = 100) -> int:
    # Parse input and create robots
    robots = parse_input(input_data)

    # Simulate movement
    final_positions = simulate_robots(robots, width, height, seconds)

    # Calculate and return safety factor
    return calculate_safety_factor(final_positions, width, height)

if __name__ == "__main__":
    with open('day14.txt', 'r') as f:
        input_data = f.read()

    result = solve_robot_safety(input_data)
    print(f"Safety factor after 100 seconds: {result}")
```

```python
import re
from dataclasses import dataclass
import numpy as np

@dataclass(slots=True)
class Robot:
    col: int
    row: int
    vcol: int
    vrow: int

def parse_input(input_data: str) -> list[Robot]:
    pattern = re.compile(r"-?\d+")
    nums = [int(m.group()) for m in re.finditer(pattern, input_data)]
    return [Robot(*nums[i:i + 4]) for i in range(0, len(nums), 4)]

def move(robots: list[Robot], seconds: int, area_rows: int, area_cols: int) -> None:
    for r in robots:
        r.row = (r.row + r.vrow * seconds) % area_rows
        r.col = (r.col + r.vcol * seconds) % area_cols

def find_tree_second(robots: list[Robot], current_seconds: int, area_rows: int, area_cols: int) -> int:
    # Calculate threshold as half of initial variance
    row_var_thres = np.var([r.row for r in robots]) * 0.5
    col_var_thres = np.var([r.col for r in robots]) * 0.5

    row_low_var, col_low_var = 0, 0

    # Find first second when variance is low in both dimensions
    while row_low_var == 0 or col_low_var == 0:
        current_seconds += 1
        move(robots, 1, area_rows, area_cols)

        if row_low_var == 0 and np.var([r.row for r in robots]) < row_var_thres:
            row_low_var = current_seconds % area_rows

        if col_low_var == 0 and np.var([r.col for r in robots]) < col_var_thres:
            col_low_var = current_seconds % area_cols

    # Find when both dimensions have low variance simultaneously
    x, y = 1, 1
    while True:
        left_side = row_low_var + area_rows * x
        right_side = col_low_var + area_cols * y
        if left_side == right_side:
            return left_side
        if left_side < right_side:
            x += 1
        else:
            y += 1

def safety_factor(robots: list[Robot], area_rows: int, area_cols: int) -> int:
    mid_row, mid_col = area_rows // 2, area_cols // 2
    quadrants = {(True, True): 0, (True, False): 0, (False, True): 0, (False, False): 0}

    for r in robots:
        if r.row == mid_row or r.col == mid_col:
            continue
        quadrants[r.row < mid_row, r.col < mid_col] += 1

    return (
        quadrants[True, True] *
        quadrants[True, False] *
        quadrants[False, True] *
        quadrants[False, False]
    )

if __name__ == "__main__":
    with open('day14.txt', 'r') as f:
        input_data = f.read()

    # Initialize robots and grid dimensions
    area_rows, area_cols = 103, 101
    robots = parse_input(input_data)

    # Solve Part 1
    move(robots, 100, area_rows, area_cols)
    print(f"Part 1: {safety_factor(robots, area_rows, area_cols)}")

    # Solve Part 2
    print(f"Part 2: {find_tree_second(robots, 100, area_rows, area_cols)}")
```

```python
from time import perf_counter_ns
from typing import Any
import os

input_file = "day15.txt"


def profiler(method):
    def wrapper_method(*args: Any, **kwargs: Any) -> Any:
        start_time = perf_counter_ns()
        ret = method(*args, **kwargs)
        duration = perf_counter_ns() - start_time
        time_len = min(9, ((len(str(duration))-1)//3)*3)
        time_conversion = {
            9: 'seconds',
            6: 'milliseconds',
            3: 'microseconds',
            0: 'nanoseconds'
        }
        print(f"Method {method.__name__} took: {duration / (10**time_len)} {time_conversion[time_len]}")
        return ret
    return wrapper_method


def free_space(robot, d, walls, boxes):
    dx, dy = d[0], d[1]

    cx, cy = robot[0] + dx, robot[1] + dy

    while True:
        if (cx, cy) in walls:
            return False, (-1, -1)
        if (cx, cy) in boxes:
            cx, cy = cx + dx, cy + dy
        else:
            return True, (cx, cy)


@profiler
def part_1():
    with open(input_file) as f:
        g, moves = f.read().split("\n\n")

    robot = (-1, -1)
    boxes = set()
    walls = set()

    deltas = {
        "^": (0, -1),
        ">": (1, 0),
        "v": (0, 1),
        "<": (-1, 0)
    }

    for y, l in enumerate(g.split("\n")):
        for x, h in enumerate(l):
            if h == "O":
                boxes.add((x, y))
            elif h == "#":
                walls.add((x, y))
            elif h == "@":
                robot = (x, y)

    moves = moves.replace("\n", "")
    for m in moves:
        dx, dy = deltas[m]
        nx, ny = robot[0] + dx, robot[1] + dy

        if (nx, ny) not in walls:
            if (nx, ny) not in boxes:
                robot = (nx, ny)
            else:
                is_free, next_free = free_space(robot, (dx, dy), walls, boxes)
                if is_free:
                    to_remove = set()
                    to_add = set()

                    for b in boxes:
                        xs = [robot[0], next_free[0]]
                        ys = [robot[1], next_free[1]]
                        if min(xs) <= b[0] <= max(xs) and min(ys) <= b[1] <= max(ys):
                            to_remove.add(b)
                            to_add.add((b[0] + dx, b[1] + dy))
                    boxes -= to_remove
                    boxes |= to_add

                    robot = (robot[0]+dx, robot[1]+dy)
    print(sum(b[0] + 100*b[1] for b in boxes))


def print_map(robot, boxes, walls):
    xs = set(w[0] for w in walls)
    ys = set(w[1] for w in walls)

    grid = [['.' for _ in range(max(xs)+1)] for _ in range(max(ys)+1)]

    for w in walls:
        grid[w[1]][w[0]] = '#'
    for b1, b2 in boxes:
        grid[b1[1]][b1[0]] = '['
        grid[b2[1]][b2[0]] = ']'

    grid[robot[1]][robot[0]] = '@'

    for y, l in enumerate(grid):
        print(f"{y:02d}", end=" ")
        print(''.join(l))


def is_free_x(robot, dx, walls, boxes):

    wx = set(w[0] for w in walls if w[1] == robot[1])
    bx = set(b[0] for bs in boxes for b in bs if b[1] == robot[1])

    cx = robot[0] + dx
    while True:
        if cx in wx:
            return False, -1
        if cx in bx:
            cx += dx
        else:
            return True, cx


def is_free_y(robot, dy, walls, boxes):
    to_move = set((b1, b2) for b1, b2 in boxes if robot[0] in [
                  b1[0], b2[0]] and robot[1] + dy == b1[1])

    top = list(to_move)
    # get the top of the stack to move
    while True:
        new_top = []
        added = False
        for b1, b2 in top:
            xs = {x: (nb1, nb2) for nb1, nb2 in boxes for x in (
                nb1[0], nb2[0]) if nb1[1] == b1[1] + dy}
            to_add = set(xs[x] for x in [b1[0], b2[0]] if x in xs)
            if len(to_add) > 0:
                new_top.extend(to_add)
                to_move |= to_add
                added = True
            else:
                new_top.append((b1, b2))

        top = new_top

        if not added:
            break

    for b1, b2 in to_move:
        if (b1[0], b1[1]+dy) in walls or (b2[0], b2[1]+dy) in walls:
            return False, None

    return True, to_move


@profiler
def part_2():
    with open(input_file) as f:
        g, moves = f.read().split("\n\n")

    robot = (-1, -1)
    boxes = set()
    walls = set()

    deltas = {
        "^": (0, -1),
        ">": (1, 0),
        "v": (0, 1),
        "<": (-1, 0)
    }

    boxes = set()
    walls = set()
    robot = (-1, -1)

    for y, l in enumerate(g.split("\n")):
        for x, h in enumerate(l):
            if h == "O":
                boxes.add(((2*x, y), (2*x+1, y)))
            elif h == "#":
                walls.add(((2*x, y)))
                walls.add(((2*x + 1, y)))
            elif h == "@":
                robot = (2*x, y)

    moves = moves.replace("\n", "")

    for m in moves:
        dx, dy = deltas[m]
        if (robot[0]+dx, robot[1]+dy) in walls:
            continue
        if dy == 0:
            is_free, next_free = is_free_x(robot, dx, walls, boxes)
            if is_free:
                min_x = min(next_free, robot[0])
                max_x = max(next_free, robot[0])
                to_move = set(
                    (b1, b2)for b1, b2 in boxes if b1[1] == robot[1] and min_x <= b1[0] <= max_x)
                boxes -= to_move
                boxes |= set(((b1[0]+dx, b1[1]), (b2[0]+dx, b2[1]))
                             for b1, b2 in to_move)
                robot = (robot[0]+dx, robot[1])
        else:
            up = set((b1, b2) for b1, b2 in boxes if robot[0] in [
                     b1[0], b2[0]] and robot[1] + dy == b1[1])
            if len(up) == 0:
                robot = (robot[0], robot[1]+dy)
            else:
                is_free, to_move = is_free_y(robot, dy, walls, boxes)

                if is_free:
                    boxes -= to_move
                    boxes |= set(((b1[0], b1[1]+dy), (b2[0], b2[1]+dy))
                                 for b1, b2 in to_move)
                    robot = (robot[0], robot[1] + dy)

    print(sum(b[0] + 100*b[1] for b, _ in boxes))


if __name__ == "__main__":
    part_1()
    part_2()
```
